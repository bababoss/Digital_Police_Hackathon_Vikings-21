```python
import geopandas as gpd
import numpy as np
from geopandas import GeoDataFrame
from shapely.geometry import Point
import folium
from folium.plugins import MarkerCluster

% store -r full_dataset
fir_data = full_dataset['tbl_CR_FIR_Details']['tbl_CR_FIR_Details']
fir_data = fir_data[fir_data['Longitude'] > 0]
fir_data = fir_data[fir_data['Latitude'] > 0]
fir_data = fir_data[fir_data['Longitude'] > 77.5081787109375]
fir_data = fir_data[fir_data['Longitude'] < 77.80682373046875]
fir_data = fir_data[fir_data['Latitude'] > 12.686662673950195]
fir_data = fir_data[fir_data['Latitude'] < 13.070075035095215]

geometry = [Point(xy) for xy in zip(fir_data.Longitude, fir_data.Latitude)]
# df = df.drop(['Lon', 'Lat'], axis=1)
crs = {'init': 'epsg:4326'}
gdf = GeoDataFrame(fir_data, crs=crs, geometry=geometry)
gdf_sample = gdf
lons = gdf_sample.geometry.x
lats = gdf_sample.geometry.y

m = folium.Map(
    location=[np.mean(lats), np.mean(lons)],
    tiles='Cartodb Positron',
    zoom_start=6
)

#FastMarkerCluster(data=list(zip(lats, lons))).add_to(m)
MarkerCluster(list(zip(lats, lons))).add_to(m)

folium.LayerControl().add_to(m)
m
```
